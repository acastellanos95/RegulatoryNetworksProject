```python
# Comenzamos la simulación
sim = [genes]
#Pasos
t=0
tEnd=N
while(t<tEnd):
    gen = [0]*N
    for i in range(N):
    #     print(index)
        entries = entriesVertex[i]
        entry1 = entries[0]
        stateEntry1 = genes[entry1]
        gen[i] = funct[i][(stateEntry1)]
    #     print(stateEntry1)
    #     print(funct[index])
    #     print(genes[index])
    #     print(funct[index][stateEntry1])
    genes = gen
    print(genes,",")
    sim.append(genes)
    t +=1
```
